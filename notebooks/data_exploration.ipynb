```python
import pandas as pd
import matplotlib.pyplot as plt
```

load the dataset

```python
df = pd.read_csv('data/pokemon.csv')
```

display the first 5 rows of the dataset

```python
print(df.head())
```

display the number of rows and columns in the dataset

```python
print(df.shape)
```

display the data types of each column

```python
print(df.dtypes)
```

display the summary statistics of the numerical columns

```python
print(df.describe())
```

display the unique values of the 'type1' column

```python
print(df['type1'].unique())
```

display the frequency count of the 'type1' column

```python
print(df['type1'].value_counts())
```

display a histogram of the 'hp' column

```python
plt.hist(df['hp'])
plt.title('Distribution of HP')
plt.xlabel('HP')
plt.ylabel('Count')
plt.show()
```

display a scatter plot of the 'attack' and 'defense' columns

```python
plt.scatter(df['attack'], df['defense'])
plt.title('Attack vs. Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()
```
