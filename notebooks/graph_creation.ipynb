```python
import pandas as pd
import networkx as nx
```

load dataset

```python
df = pd.read_csv('pokemon.csv')
```

create empty graph

```python
G = nx.Graph()
```

add nodes to graph

```python
for index, row in df.iterrows():
    G.add_node(row['name'], type1=row['type1'], type2=row['type2'])
```

add edges to graph

```python
for index, row in df.iterrows():
    for col in df.columns:
        if col.startswith('against_') and row[col] > 1:
            type_attacked = col.split('against_')[1]
            for node in G.nodes(data=True):
                if node[1]['type1'] == type_attacked or node[1]['type2'] == type_attacked:
                    G.add_edge(row['name'], node[0])
```
